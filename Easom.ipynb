# Easom

Easom is interesting, it has a very deep optimum and is relatively flat elsewhere.

Reference: https://www.sfu.ca/~ssurjano/easom.html

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Function Definition

```python
lowBound = -100
highBound = 100
ranges = []
for i in range(2):
    ranges.append([-lowBound, highBound])
#Note: global minimum is f(x*) = -1, x* = (pi, pi) for standard easom

def easom(x):
    return(-np.cos(x[0]) * np.cos(x[1]) * np.e ** sum(-((c - np.pi) ** 2) for c in x))
```

## Plotting

```python
resolution = 1
inputs = []
for x in np.arange(lowBound, highBound, resolution):
    for y in np.arange(lowBound, highBound, resolution):
        inputs.append([x, y])
        
xs = [point[0] for point in inputs]
ys = [point[1] for point in inputs]
cs = [easom(point) for point in inputs]

plt.title("Easom Heatmap")
plt.scatter(xs, ys, c = cs, cmap = "viridis") 
plt.colorbar(label = "Easom value")
plt.xlim(lowBound, highBound)
plt.ylim(lowBound, highBound)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
