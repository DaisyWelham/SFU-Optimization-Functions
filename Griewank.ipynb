# Griewank

Griewank is an interesting optimization function. It has a large number of local minima, which can cause algorithms to become trapped in local minima. It can also be defined on any number of dimensions, which makes it an interesting function for high-dimensional optimization problems.

Reference: https://www.sfu.ca/~ssurjano/griewank.html

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Function Definition

```python
lowBound = -600
highBound = 600
numDimensions = 2 #Usually 10+, but 2 is better for plotting
divider = 4000 #4000 in standard griewank, but in principle we can change this
constantTerm = 1 #1 makes the global optimum f(x*) = 0, which is usually convenient. 
ranges = []
for i in range(numDimensions):
    ranges.append([-lowBound, highBound])
#Note: global minimum is f(x*) = 0, x* = (0, ... 0) for standard griewank

def griewank(x):
    sumTerm = 0
    productTerm = 1
    for i in range(len(x)):
        value = x[i]
        sumTerm += (value ** 2)/ divider
        productTerm *= np.cos(value / ((i + 1) ** 0.5))
    return(sumTerm - productTerm + constantTerm)
```

## Plotting

```python
resolution = 10
inputs = []
for x in np.arange(lowBound, highBound, resolution):
    for y in np.arange(lowBound, highBound, resolution):
        inputs.append([x, y])
        
xs = [point[0] for point in inputs]
ys = [point[1] for point in inputs]
cs = [griewank(point) for point in inputs]

plt.title("Griewank Heatmap")
plt.scatter(xs, ys, c = cs, cmap = "coolwarm") 
plt.colorbar(label = "Griewank value")
plt.xlim(lowBound, highBound)
plt.ylim(lowBound, highBound)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
