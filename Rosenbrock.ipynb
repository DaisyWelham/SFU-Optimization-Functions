# Rosenbrock

Rosenbrock has a banana-shaped valley inside the main valley. This makes Rosenbrock interesting because it is effectively a threefold problem: can you find the valley? If you find the valley, can you find the banana? If you find the banana, can you find the global optimum?

Reference: https://www.sfu.ca/~ssurjano/Rosenbrock.html

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Function Definition

```python
lowBound = -10
highBound = 10
numDimensions = 2 #Usually 10+, but 2 is better for plotting. Must be at least 2.
constant = 100 #The constant in front of the xn+1 - xn term
ranges = []
for i in range(numDimensions):
    ranges.append([-lowBound, highBound])
#Note: global minimum is f(x*) = 0, x* = (1, ... 1) for standard Rosenbrock

def rosenbrock(x):
    output = 0
    for i in range(len(x) - 1):
        xPlus = x[i + 1]
        output += 100 * (xPlus - (x[i]) ** 2) ** 2 + (x[i] - 1) ** 2
    return(output)
```

## Plotting

```python
resolution = 0.1
inputs = []
for x in np.arange(lowBound, highBound, resolution):
    for y in np.arange(lowBound, highBound, resolution):
        inputs.append([x, y])
        
xs = [point[0] for point in inputs]
ys = [point[1] for point in inputs]
cs = [rosenbrock(point) for point in inputs]

plt.title("Rosenbrock Heatmap")
plt.scatter(xs, ys, c = cs, cmap = "viridis") 
plt.colorbar(label = "Rosenbrock value")
plt.xlim(lowBound, highBound)
plt.ylim(lowBound, highBound)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

## Close Up

```python
resolution = 0.01
lowBound = -1.25
highBound = 1.25
inputs = []
for x in np.arange(lowBound, highBound, resolution):
    for y in np.arange(lowBound, highBound, resolution):
        inputs.append([x, y])
        
xs = [point[0] for point in inputs]
ys = [point[1] for point in inputs]
cs = [rosenbrock(point) for point in inputs]

plt.title("Rosenbrock Heatmap (Banana)")
plt.scatter(xs, ys, c = cs, cmap = "viridis") 
plt.colorbar(label = "Rosenbrock value")
plt.xlim(lowBound, highBound)
plt.ylim(lowBound, highBound)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
