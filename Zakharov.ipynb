# Zakharov

Zakharov is interesting because there is only one local minimum, which of course is therefore also the global minimum. It can be defined on a large number of dimensions, which makes it useful for a wide range of research problems.

Reference: https://www.sfu.ca/~ssurjano/zakharov.html

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Function Definition

```python
lowBound = -5
highBound = 10
numDimensions = 2 #Usually 10+, but 2 is better for plotting
squareConstant = 0.5 #In case we want to play with parameters- the constant in fromt of ix^2
quarticConstant = 0.5 #In case we want to play with parameters- the constant in fromt of ix^4
ranges = []
for i in range(numDimensions):
    ranges.append([-lowBound, highBound])
#Note: global minimum is f(x*) = 0, x* = (0, ... 0) for standard zakharov

def zakharov(x):
    sum1 = sum(val ** 2 for val in x)

    sum2 = sum(squareConstant * (i + 1) * x[i] for i in range(len(x)))
    sum2 = sum2 ** 2

    sum3 = sum(quarticConstant * (i + 1) * x[i] for i in range(len(x)))
    sum3 = sum3 ** 4

    return(sum1 + sum2 + sum3)
```

## Plotting

```python
resolution = 0.1
inputs = []
for x in np.arange(lowBound, highBound, resolution):
    for y in np.arange(lowBound, highBound, resolution):
        inputs.append([x, y])
        
xs = [point[0] for point in inputs]
ys = [point[1] for point in inputs]
cs = [zakharov(point) for point in inputs]

plt.title("Zakharov Heatmap")
plt.scatter(xs, ys, c = cs, cmap = "viridis") 
plt.colorbar(label = "Zakharov value")
plt.xlim(lowBound, highBound)
plt.ylim(lowBound, highBound)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

## Close Up

```python
resolution = 0.01
lowBound = -1
highBound = 1
inputs = []
for x in np.arange(lowBound, highBound, resolution):
    for y in np.arange(lowBound, highBound, resolution):
        inputs.append([x, y])
        
xs = [point[0] for point in inputs]
ys = [point[1] for point in inputs]
cs = [zakharov(point) for point in inputs]

plt.title("Zakharov Heatmap")
plt.scatter(xs, ys, c = cs, cmap = "viridis") 
plt.colorbar(label = "Zakharov value")
plt.xlim(lowBound, highBound)
plt.ylim(lowBound, highBound)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
