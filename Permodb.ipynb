# Perm-0-d-beta

Perm-0-d-beta ("podb") resembles a steep "bowl" and can be implemented in a large number of dimensions.

References: https://www.sfu.ca/~ssurjano/perm0db.html

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Function Definition

```python
numDimensions = 2 #In principle can be large, but 2 is better for plotting
beta = 0.1 #We can tune this to change the steepness of podb
ranges = []
for i in range(numDimensions):
    ranges.append([-numDimensions, numDimensions])
#Note: global minimum is f(x*) = 0, x* = (1, 1/2, ..., 1/d) for standard podb

def podb(x):
    output = 0
    for I in range(len(x)):
        subSum = 0
        for J in range(len(x)):
            i = I + 1
            j = J + 1
            subSum += (j + beta) * (x[J] ** i - (j ** -i))
        output += subSum ** 2
    return(output)
```

## Plotting

```python
resolution = 0.1
inputs = []
for x in np.arange(-numDimensions, numDimensions, resolution):
    for y in np.arange(-numDimensions, numDimensions, resolution):
        inputs.append([x, y])
        
xs = [point[0] for point in inputs]
ys = [point[1] for point in inputs]
cs = [podb(point) for point in inputs]

plt.title("Perm0db Heatmap")
plt.scatter(xs, ys, c = cs, cmap = "viridis") 
plt.colorbar(label = "Perm0db value")
plt.xlim(-numDimensions, numDimensions)
plt.ylim(-numDimensions, numDimensions)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
