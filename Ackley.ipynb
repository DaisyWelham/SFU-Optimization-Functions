# Ackley

Reference: https://www.sfu.ca/~ssurjano/ackley.html

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Function Definition

```python
lowBound = -32.768
highBound = 32.768
numDimensions = 2 #Usually 10+, but 2 is better for plotting
a = 20
b = 0.2
c = 2 * np.pi
ranges = []
for i in range(numDimensions):
    ranges.append([-lowBound, highBound])
#Note: global minimum is f(x*) = 0, x* = (0, ... 0) for standard ackley

def ackley(x):
    sum1 =  - b * ((1 / numDimensions) * sum(k ** 2 for k in x)) ** 0.5
    sum2 = (1 / numDimensions) * sum(np.cos(c * k) for k in x)
    return(-a * (np.e ** sum1) - (np.e ** sum2) + a + np.e)
```

## Plotting

```python
resolution = 0.1
inputs = []
for x in np.arange(lowBound, highBound, resolution):
    for y in np.arange(lowBound, highBound, resolution):
        inputs.append([x, y])
        
xs = [point[0] for point in inputs]
ys = [point[1] for point in inputs]
cs = [ackley(point) for point in inputs]

plt.title("Ackley Heatmap")
plt.scatter(xs, ys, c = cs, cmap = "coolwarm") 
plt.colorbar(label = "Ackley value")
plt.xlim(lowBound, highBound)
plt.ylim(lowBound, highBound)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
