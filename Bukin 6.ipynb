# Bukin 6

Bukin 6 is not differentiable which messes up gradient-based algorithms like gradient descent. The local optima that lie on the ridge can also cause local search algorithms to get trapped, 

Reference: https://www.sfu.ca/~ssurjano/bukin6.html

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Function Definition

```python
numDimensions = 2 #Usually 10+, but 2 is better for plotting
a = 20
b = 0.2
c = 2 * np.pi
ranges = [[-15, -5],[-3, 3]]
#Note: global minimum is f(x*) = 0, x* = (-10, 1)

def bukin(x):
    return(100 * np.fabs(x[1] - 0.01 * (x[0]) ** 2) ** 0.5 + 0.01 * np.fabs(x[0] + 10))
```

## Plotting

```python
resolution = 0.05
inputs = []
for x in np.arange(-15, -5, resolution):
    for y in np.arange(-3, 3, resolution):
        inputs.append([x, y])
        
xs = [point[0] for point in inputs]
ys = [point[1] for point in inputs]
cs = [bukin(point) for point in inputs]

plt.title("Bukin6 Heatmap")
plt.scatter(xs, ys, c = cs, cmap = "coolwarm") 
plt.colorbar(label = "Bukin6 value")
plt.xlim(-15, -5)
plt.ylim(-3, 3)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
