# Dixon-Price

Reference: https://www.sfu.ca/~ssurjano/dixonpr.html

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Function Definition

```python
lowBound = -10
highBound = 10
numDimensions = 2 #Usually 10+, but 2 is better for plotting
ranges = []
for i in range(numDimensions):
    ranges.append([-lowBound, highBound])
#Note: global minimum is f(x*) = 0 at x* = xi for i in 2 ^ - {2^i - 2}{2^i}

def dixonPrice(x):
    output = (x[0] - 1) ** 2
    for j in range(numDimensions - 1):
        i = j + 1
        output += i * ((2 * (x[i]) ** 2) - x[i - 1]) ** 2
    return(output)
```

## Plotting

```python
resolution = 0.1
inputs = []
for x in np.arange(lowBound, highBound, resolution):
    for y in np.arange(lowBound, highBound, resolution):
        inputs.append([x, y])
        
xs = [point[0] for point in inputs]
ys = [point[1] for point in inputs]
cs = [dixonPrice(point) for point in inputs]

plt.title("Dixon-Price Heatmap")
plt.scatter(xs, ys, c = cs, cmap = "coolwarm") 
plt.colorbar(label = "Dixon-Price value")
plt.xlim(lowBound, highBound)
plt.ylim(lowBound, highBound)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
