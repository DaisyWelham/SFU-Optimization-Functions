# Dropwave

Dropwave has many local minima and it looks like the waves resuling from dropping a rock in some water. The many local optima cause problems for many kinds of optimization algorithms.

Reference: https://www.sfu.ca/~ssurjano/drop.html

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Function Definition

```python
lowBound = -5.12
highBound = 5.12
ranges = []
for i in range(2):
    ranges.append([-lowBound, highBound])

def dropwave(x):
    return( -((1 + np.cos(12 * (x[0] ** 2 + x[1] ** 2) ** 0.5)) / (0.5 * (x[0] ** 2 + x[1] ** 2) + 2)))
#Optimum f(x*) = -1 at x* = (0, 0)
```

## Plotting

```python
resolution = 0.1
inputs = []
for x in np.arange(lowBound, highBound, resolution):
    for y in np.arange(lowBound, highBound, resolution):
        inputs.append([x, y])
        
xs = [point[0] for point in inputs]
ys = [point[1] for point in inputs]
cs = [dropwave(point) for point in inputs]

plt.title("Dropwave Heatmap")
plt.scatter(xs, ys, c = cs, cmap = "coolwarm") 
plt.colorbar(label = "Dropwave value")
plt.xlim(lowBound, highBound)
plt.ylim(lowBound, highBound)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
